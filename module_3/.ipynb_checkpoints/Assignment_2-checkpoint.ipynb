```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('cereal.csv')
data.head()
```

```python
plt.hist(data[['sugars','vitamins']])
plt.legend(['sugars','vitamins'])
plt.show()
```

```python
mapping ={ 'N': 'Nabisco','Q': 'Quaker Oats','K': 'Kelloggs','R': 'Raslston Purina','G': 'General Mills' ,'P' :'Post' ,'A':'American Home Foods Products'}
```

```python
data['Manufacturer']=data['mfr'].map(mapping)
data.head()
```

```python
df = data[['name','Manufacturer']].groupby(['Manufacturer']).agg(['count'])
df.shape
```

```python
data[['name','Manufacturer']].groupby(['Manufacturer']).count()
#data['Manufacturer'].value_counts
#data.apply(pd.value_counts).fillna(0)
```
