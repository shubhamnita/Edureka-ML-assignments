# 2.Data Manipulation:

```python
data['total_benefitted'] = data.sum(axis=1)
data.head()
```

```python
data.loc['totals']= data.sum(numeric_only=True, axis=0)
data.at['totals','YEAR'] =2013
data
```

# 3.Plotting:

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = data.dropna()
plt.bar(df['STATE/UT'],df['total_benefitted'],align='center')
plt.xlabel('States/UT')
plt.ylabel('Total beneficaries')
plt.show()
```
